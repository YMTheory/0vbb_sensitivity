```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('sensitivity.mplstyle')
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
sens_energy_res = np.genfromtxt('Data/d024_sensitivity_vs_energy_resolution.txt',\
                                delimiter=',',skip_header=1)

baseline_sens = sens_energy_res[0,1]/1.e28 # x10^28 yrs

```

```python
cut = sens_energy_res[:,0] < 1.21

plt.plot(sens_energy_res[cut,0],sens_energy_res[cut,1]/1.e28,\
         'o',markersize=7,color=colors[0],label='90\% CL upper limit')
plt.xlim(.7,1.3)
plt.ylim(1.,1.5)
plt.ylabel(r'Halflife ($\times 10^{28}$ yrs)')
plt.xlabel(r'$\sigma/E$ energy resolution at $Q_{\beta\beta}$ (\%)')
plt.grid()

p = np.polyfit(sens_energy_res[cut,0],sens_energy_res[cut,1]/1.e28,1)
print(p)

xfit = np.linspace(0.7,1.3,100)
yfit = p[0]*xfit + p[1]

plt.plot(xfit,yfit,'-',color=colors[0],linewidth=3)

plt.plot(0.8,baseline_sens,\
           'o',markersize=14,linewidth=5,color=colors[0])
plt.legend(loc='lower left',fontsize=16)

plt.savefig('sensitivity_vs_energy_resolution_v11.png',dpi=300,bbox_inches='tight')

```
