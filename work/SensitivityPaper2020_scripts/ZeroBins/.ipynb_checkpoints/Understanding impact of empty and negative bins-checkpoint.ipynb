```python
zeros_in_sum = sumhist.values == 0.
print(np.sum(zeros_in_sum))

zeros_in_sig = sighist.values==0.
print(np.sum(zeros_in_sig))

print('Zeros in sum but not in sig: {}'.format(np.sum( zeros_in_sum & np.invert(zeros_in_sig) )))


```

```python
sumhistnew = sumhist - 1.0 * sighist



zeros_in_sum = (sumhistnew.values == 0.) | (sumhistnew.values < 0.)
print(np.sum(zeros_in_sum))

zeros_in_sig = sighist.values==0.
print(np.sum(zeros_in_sig))

print('Zeros in sum but not in sig: {}'.format(np.sum( zeros_in_sum & np.invert(zeros_in_sig) )))

badbins = zeros_in_sum & np.invert(zeros_in_sig)

print(badbins)


```

```python
badhist = hl.Hist(sumhistnew.bins,badbins.astype(float))

hl.plot2d(badhist.project([1,2]),log=True)

plt.figure(2)
hl.plot2d(sumhist.project([1,2]),log=True)
#hl.plot2d(sumhist,log=True)
```
